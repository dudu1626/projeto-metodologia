```python
sns.heatmap(dados_cifar_treino[['duration', 'emissions', 'emissions_rate', 'cpu_power', 'gpu_power', 'ram_power', 'cpu_energy', 'gpu_energy', 'ram_energy', 'energy_consumed']].corr(), annot=True)
plt.title('Matriz de correlação', fontsize=16)
plt.show()
```

```python
# Gráfico
sns.pairplot(dados_cifar_treino[['Modelo', 'duration', 'emissions', 'emissions_rate', 'cpu_power', 'gpu_power', 'ram_power', 'cpu_energy', 'gpu_energy', 'ram_energy', 'energy_consumed']], hue='Modelo')
plt.title('Gráfico de pares', fontsize=16)
plt.show()
```

```python
sns.boxplot(x='Modelo', y='duration', data=dados_cifar_treino)
plt.title('Boxplot de duração', fontsize=16)
plt.show()
```

```python
sns.boxplot(x='Modelo', y='emissions', data=dados_cifar_treino)
plt.title('Boxplot de emissões', fontsize=16)
plt.show()
```

```python
# sns.histplot(dados_cifar_treino, x='emissions', hue='Modelo', multiple='stack', kde=True)
sns.histplot(dados_cifar_treino, x='emissions', hue='Modelo', multiple='stack')

plt.title('Histograma de emissões', fontsize=16, fontweight='bold')
plt.show()
```
